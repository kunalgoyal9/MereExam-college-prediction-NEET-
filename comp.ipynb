```python
#importing various libraries needed for our model

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
import re
import matplotlib.pyplot as plt
```

```python
all_india = pd.read_csv('round_1_alloted.csv', sep=',', header=None)
```

```python
deemed = pd.read_csv('tabula-deemed_college.csv', sep=',', header=None)
```

```python
all_india = all_india[2]
all_india.head()
```

```python
deemed = deemed[3]
deemed = deemed.dropna()
deemed.head()
```

```python
colleges = pd.concat([all_india, deemed], axis=0)
colleges = colleges.str.lower()
```

```python
import re
def correct(clg):
    clg = re.sub(r"\r"," ", clg)
    clg = re.sub(r"\. ", ".", clg)
    clg = re.sub(r"university", "univ.", clg)
    clg = re.sub(r"hospital", "hosp.", clg)
    return re.sub(r"\n"," ", clg)
```

```python
def cor(a):
    return re.subn(r"\. ",".", a)
```

```python
cor('hello. man. ha')[0]
```

```python
colleges = colleges.apply(correct)
```

```python
colleges = pd.DataFrame(list(set(list(colleges))), columns=['Colleges'])
colleges = colleges.sort_values(by=['Colleges'])
colleges
```

```python
vibhu_d = pd.read_csv('cid_courseid_mbbs.csv')
vibhu_d['name'] = vibhu_d['name'].str.lower()
vibhu_d['name'] = vibhu_d['name'].apply(correct)
vibhu_d = vibhu_d.sort_values(by= ['name'])
vibhu_d
```
